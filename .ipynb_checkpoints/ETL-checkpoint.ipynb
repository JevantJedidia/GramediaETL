```python
import requests
import pandas as pd
from pandas import json_normalize
import json
import random
from datetime import datetime, timedelta
```

```python
## Tentukan range waktu untuk generate timestamp
start = datetime(2025, 10, 1)
end = datetime(2025, 10, 31, 23, 59, 59)
```

```python
## sementara dari local
# r_product = requests.get('https://fakestoreapi.com/products')
# r_sales = requests.get('https://fakestoreapi.com/carts')

## Load data products
with open('products.json', 'r') as file:
    r_product = json.load(file)

df_product = pd.concat(json_normalize(r_product['products']) for i in range (len(r_product['products'])))

## Load data carts
with open('carts.json', 'r') as file:
    r_carts = json.load(file)

df_sales = pd.DataFrame()
for item in r_carts['carts']:
    df_temp = json_normalize(item['products'])
    df_temp['transactiondate'] = start + (end - start) * random.random()
    df_sales = pd.concat([df_sales,df_temp],ignore_index=True)
```

```python
df_product = df_product[['id','title','description','category','price']]
df_sales = df_sales[['id','title','price','quantity','transactiondate']]
```

```python
df_merge = pd.merge(df_product,df_sales,on='title',how='inner')
df_merge = df_merge[['id_x','title','category','quantity','price_y','transactiondate']].rename(columns ={
        'id_x': 'product_id',
        'title': 'product_name',
        'price_y': 'price', 
    }
)
df_merge['total_sales'] = df_merge['quantity']*df_merge['price']
```
