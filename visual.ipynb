```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Connect to SQlite
conn = sqlite3.connect("Ecommerce.db")

## definte query
query1 = "SELECT substr(transaction_date,1,10) date, sum(total_sales) total_sales FROM sales group by substr(transaction_date,1,10)"
df1 = pd.read_sql_query(query1,conn)

query2 = "SELECT b.category, sum(a.total_sales) total_sales from sales a join category b on a.category = b.category_id group by b.category"
df2 = pd.read_sql_query(query2,conn)
conn.close()
```

```python
## Visualisasi total_sales per date
df1.plot(kind='bar', x='date', y='total_sales', legend=False)
plt.title("Total Sales per Date")
plt.show()
```

```python
## Visualisasi total_sales per category
df2.plot(kind='bar', x='category', y='total_sales', legend=False)
plt.title("Total Sales per Category")
plt.show()
```
